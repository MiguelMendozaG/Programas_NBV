```python
#crea un octree de 32x32x32 del archivo de entrada indicado
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
f11_r3d = np.array([36.46,68.07,74.76,94.58,94.79,94.97,95.61])
f11_vn = np.array([36.46,73.71,90.96,92.07])
print(len(f11_r3d))
```

```python
fig = plt.figure()
fig, ax = plt.subplots()
x_r3d = np.arange(0, len(f11_r3d),1)
x_vn = np.arange(0, len(f11_vn),1)
ax.set_title("Figure 1")
ax.set_xlabel('Iteration')
ax.set_ylabel('% of covered surface')
ax.plot(x_r3d, f11_r3d, 'b--', label = 'Proposed network')
ax.plot(x_vn, f11_vn, 'r', label = 'Voxnet')
ax.legend()
fig.show()
fig.savefig("cov_surf_f1.jpg", dpi = 300)
```
